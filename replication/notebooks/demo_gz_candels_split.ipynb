```python
import shutil
import os

import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_parquet('/Users/user/repos/galaxy-datasets/roots/gz_candels/candels_ortho_train_catalog.parquet')  # can be found on dropbox
df
```

```python
df = df.sample(1000, random_state=42)
```

```python
df['filename']
```

```python
for loc in df['filename']:
    original_loc = os.path.join('/Users/user/repos/galaxy-datasets/roots/gz_candels/images', loc)
    new_loc = os.path.join('/Users/user/repos/galaxy-datasets/data/derived_data/demo_gz_candels/images', loc)
    shutil.copyfile(original_loc, new_loc)
```

```python
# tar -czvf demo_gz_candels_images.tar.gz images
```

```python
train_catalog, test_catalog = train_test_split(df, test_size=0.2, random_state=42)
```

```python
train_catalog_loc = '/Users/user/repos/galaxy-datasets/roots/demo_gz_candels/demo_gz_candels_train_catalog.parquet'
test_catalog_loc = '/Users/user/repos/galaxy-datasets/roots/demo_gz_candels/demo_gz_candels_test_catalog.parquet'

train_catalog.to_parquet(train_catalog_loc, index=False)
test_catalog.to_parquet(test_catalog_loc, index=False)
```

```python
import hashlib

for loc in [
    train_catalog_loc,
    test_catalog_loc,
    '/Users/user/repos/galaxy-datasets/data/derived_data/demo_gz_candels/demo_gz_candels_images.tar.gz'
]:
    # print hash
    with open(loc, 'rb') as f:
        md5_checksum = hashlib.md5(f.read()).hexdigest()

    print(md5_checksum)
```
